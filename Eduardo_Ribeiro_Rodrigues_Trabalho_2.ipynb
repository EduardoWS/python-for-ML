Para calcular a probabilidade de que o pre√ßo m√©dio da amostra esteja entre R$18,00 e R$25,00, podemos usar a distribui√ß√£o normal e calcular o intervalo de confian√ßa.

O intervalo de confian√ßa √© calculado usando a f√≥rmula:

    Intervalo de confian√ßa = m√©dia da amostra ¬± (valor cr√≠tico * erro padr√£o)

O valor cr√≠tico depende do n√≠vel de confian√ßa desejado. Para este exemplo, vamos considerar um n√≠vel de confian√ßa de 95%, o que corresponde a um valor cr√≠tico de 1,96 (para um tamanho de amostra maior que 30).

O erro padr√£o √© calculado dividindo o desvio padr√£o pela raiz quadrada do tamanho da amostra:

    Erro padr√£o = desvio padr√£o / sqrt(tamanho da amostra)

Vamos calcular o intervalo de confian√ßa em Python:

```python
import math

# Dados do problema
media_populacao = 24.5
desvio_padrao_populacao = 5.45
tamanho_amostra = 50
nivel_confianca = 0.95
valor_critico = 1.96

# C√°lculo do erro padr√£o
erro_padrao = desvio_padrao_populacao / math.sqrt(tamanho_amostra)

# C√°lculo do intervalo de confian√ßa
intervalo_inferior = media_populacao - valor_critico * erro_padrao
intervalo_superior = media_populacao + valor_critico * erro_padrao

# Verifica√ß√£o se o pre√ßo m√©dio da amostra est√° entre R$18,00 e R$25,00
if intervalo_inferior <= 18.0 and intervalo_superior >= 25.0:
    probabilidade = 1.0
elif intervalo_inferior <= 18.0:
    probabilidade = (intervalo_superior - 18.0) / (intervalo_superior - intervalo_inferior)
elif intervalo_superior >= 25.0:
    probabilidade = (25.0 - intervalo_inferior) / (intervalo_superior - intervalo_inferior)
else:
    probabilidade = 0.0

print("Probabilidade:", probabilidade)
```

Para calcular a probabilidade de que uma amostra aleat√≥ria de 36 resistores el√©tricos tenha uma resist√™ncia combinada de mais de 1458 ohms, precisamos usar a distribui√ß√£o normal e calcular a m√©dia e o desvio padr√£o da resist√™ncia combinada.

A resist√™ncia combinada de uma amostra de resistores √© a soma das resist√™ncias individuais de cada resistor na amostra. Como estamos assumindo que os resistores s√£o independentes, podemos usar as seguintes propriedades:

    M√©dia da resist√™ncia combinada = N√∫mero de resistores na amostra * M√©dia da resist√™ncia individual

    Desvio padr√£o da resist√™ncia combinada = sqrt(N√∫mero de resistores na amostra) * Desvio padr√£o da resist√™ncia individual

Vamos calcular a probabilidade em Python:

```python
import math

# Dados do problema
media_resistencia = 40.0
desvio_padrao_resistencia = 2.0
tamanho_amostra = 36
resistencia_desejada = 1458.0

# C√°lculo da m√©dia e desvio padr√£o da resist√™ncia combinada
media_combinada = tamanho_amostra * media_resistencia
desvio_padrao_combinado = math.sqrt(tamanho_amostra) * desvio_padrao_resistencia

# C√°lculo da probabilidade usando a distribui√ß√£o normal
z_score = (resistencia_desejada - media_combinada) / desvio_padrao_combinado
probabilidade = 1.0 - 0.5 * (1.0 + math.erf(z_score / math.sqrt(2)))

print("Probabilidade:", probabilidade)
```

Para construir o intervalo com 90% de confian√ßa para a m√©dia populacional, podemos usar a distribui√ß√£o t-Student, uma vez que o tamanho da amostra √© menor do que 30 e o desvio padr√£o populacional n√£o √© conhecido.

O intervalo de confian√ßa √© calculado usando a f√≥rmula:

    Intervalo de confian√ßa = m√©dia da amostra ¬± (valor cr√≠tico * erro padr√£o)

O valor cr√≠tico depende do n√≠vel de confian√ßa desejado e do tamanho da amostra menos um. Para um n√≠vel de confian√ßa de 90% e um tamanho de amostra de 80-1=79, o valor cr√≠tico pode ser obtido usando uma tabela t-Student ou uma fun√ß√£o de distribui√ß√£o t em Python.

O erro padr√£o √© calculado dividindo o desvio padr√£o pelo valor da raiz quadrada do tamanho da amostra:

    Erro padr√£o = desvio padr√£o / sqrt(tamanho da amostra)

Vamos calcular o intervalo de confian√ßa em Python:

```python
import math
from scipy.stats import t

# Dados do problema
media_amostra = 4.8
desvio_padrao_populacao = 2.2
tamanho_amostra = 80
nivel_confianca = 0.9

# C√°lculo do valor cr√≠tico
graus_de_liberdade = tamanho_amostra - 1
valor_critico = t.ppf((1 + nivel_confianca) / 2, graus_de_liberdade)

# C√°lculo do erro padr√£o
erro_padrao = desvio_padrao_populacao / math.sqrt(tamanho_amostra)

# C√°lculo do intervalo de confian√ßa
intervalo_inferior = media_amostra - valor_critico * erro_padrao
intervalo_superior = media_amostra + valor_critico * erro_padrao

print("Intervalo de confian√ßa:", intervalo_inferior, "-", intervalo_superior)
```

A sa√≠da √© o intervalo de confian√ßa de 90% para a m√©dia populacional.

Podemos calcular os intervalos de confian√ßa para a capacidade m√©dia das baterias usando a distribui√ß√£o t-Student.

Para construir o intervalo de confian√ßa, utilizamos a f√≥rmula:

    Intervalo de confian√ßa = m√©dia da amostra ¬± (valor cr√≠tico * erro padr√£o)

O valor cr√≠tico depende do n√≠vel de confian√ßa desejado e do tamanho da amostra menos um. Para um n√≠vel de confian√ßa de 95% ou 99%, e um tamanho de amostra de 121-1=120, podemos obter os valores cr√≠ticos utilizando uma tabela t-Student ou uma fun√ß√£o de distribui√ß√£o t em Python.

O erro padr√£o √© calculado dividindo o desvio padr√£o pela raiz quadrada do tamanho da amostra:

    Erro padr√£o = desvio padr√£o / sqrt(tamanho da amostra)

Vamos calcular os intervalos de confian√ßa em Python:

```python
import math
from scipy.stats import t

# Dados do problema
media_amostra = 178
desvio_padrao_amostra = 14
tamanho_amostra = 121
nivel_confianca_95 = 0.95
nivel_confianca_99 = 0.99

# C√°lculo dos valores cr√≠ticos
graus_de_liberdade = tamanho_amostra - 1
valor_critico_95 = t.ppf((1 + nivel_confianca_95) / 2, graus_de_liberdade)
valor_critico_99 = t.ppf((1 + nivel_confianca_99) / 2, graus_de_liberdade)

# C√°lculo do erro padr√£o
erro_padrao = desvio_padrao_amostra / math.sqrt(tamanho_amostra)

# C√°lculo dos intervalos de confian√ßa
intervalo_inferior_95 = media_amostra - valor_critico_95 * erro_padrao
intervalo_superior_95 = media_amostra + valor_critico_95 * erro_padrao

intervalo_inferior_99 = media_amostra - valor_critico_99 * erro_padrao
intervalo_superior_99 = media_amostra + valor_critico_99 * erro_padrao

print("Intervalo de confian√ßa 95%:", intervalo_inferior_95, "-", intervalo_superior_95)
print("Intervalo de confian√ßa 99%:", intervalo_inferior_99, "-", intervalo_superior_99)
```

A sa√≠da √© os intervalos de confian√ßa de 95% e 99% para a capacidade m√©dia das baterias produzidas pelo m√©todo.

Para determinar o intervalo de confian√ßa para a porcentagem de consumidores do produto, podemos usar a distribui√ß√£o binomial.

O intervalo de confian√ßa √© calculado usando a f√≥rmula:

    Intervalo de confian√ßa = propor√ß√£o da amostra ¬± (valor cr√≠tico * erro padr√£o)

O valor cr√≠tico depende do n√≠vel de confian√ßa desejado. Para um n√≠vel de confian√ßa de 95%, usamos um valor cr√≠tico z de aproximadamente 1,96.

O erro padr√£o √© calculado dividindo a raiz quadrada da propor√ß√£o amostral pelo tamanho da amostra:

    Erro padr√£o = sqrt((propor√ß√£o amostral * (1 - propor√ß√£o amostral)) / tamanho da amostra)

Vamos calcular o intervalo de confian√ßa em Python:

```python
import math

# Dados do problema
tamanho_amostra = 300
consumidores = 100
nivel_confianca = 0.95

# C√°lculo da propor√ß√£o amostral
proporcao_amostral = consumidores / tamanho_amostra

# C√°lculo do valor cr√≠tico
valor_critico = 1.96

# C√°lculo do erro padr√£o
erro_padrao = math.sqrt((proporcao_amostral * (1 - proporcao_amostral)) / tamanho_amostra)

# C√°lculo do intervalo de confian√ßa
intervalo_inferior = proporcao_amostral - valor_critico * erro_padrao
intervalo_superior = proporcao_amostral + valor_critico * erro_padrao

print("Intervalo de confian√ßa:", intervalo_inferior, "-", intervalo_superior)
```

A sa√≠da √© o intervalo de confian√ßa para a porcentagem de consumidores do produto. Por exemplo, se o intervalo for de 0,25 a 0,35, isso significa que com 95% de confian√ßa, a porcentagem real de consumidores do produto est√° entre 25% e 35%.

Para construir o intervalo de confian√ßa para Œº (m√©dia) e œÉ¬≤ (vari√¢ncia), podemos usar a distribui√ß√£o t-Student e a distribui√ß√£o qui-quadrado, respectivamente.

**(a) Intervalo de confian√ßa de 95% para Œº:**

Podemos calcular o intervalo de confian√ßa para a m√©dia usando a seguinte f√≥rmula:

    Intervalo de confian√ßa = m√©dia da amostra ¬± (valor cr√≠tico * erro padr√£o)

O valor cr√≠tico depende do n√≠vel de confian√ßa desejado e do tamanho da amostra menos um. Para um n√≠vel de confian√ßa de 95% e um tamanho de amostra de 10-1=9, podemos obter o valor cr√≠tico usando uma tabela t-Student ou uma fun√ß√£o de distribui√ß√£o t em Python.

O erro padr√£o √© calculado dividindo o desvio padr√£o pela raiz quadrada do tamanho da amostra:

    Erro padr√£o = desvio padr√£o / sqrt(tamanho da amostra)

Vamos calcular o intervalo de confian√ßa para Œº em Python:

```python
import math
from scipy.stats import t

# Dados do problema
amostra = [9, 8, 12, 7, 9, 6, 11, 6, 10, 9]
tamanho_amostra = len(amostra)
nivel_confianca = 0.95

# C√°lculo da m√©dia e desvio padr√£o da amostra
media_amostra = sum(amostra) / tamanho_amostra
desvio_padrao_amostra = math.sqrt(sum((x - media_amostra) ** 2 for x in amostra) / (tamanho_amostra - 1))

# C√°lculo do valor cr√≠tico
graus_de_liberdade = tamanho_amostra - 1
valor_critico = t.ppf((1 + nivel_confianca) / 2, graus_de_liberdade)

# C√°lculo do erro padr√£o
erro_padrao = desvio_padrao_amostra / math.sqrt(tamanho_amostra)

# C√°lculo do intervalo de confian√ßa
intervalo_inferior = media_amostra - valor_critico * erro_padrao
intervalo_superior = media_amostra + valor_critico * erro_padrao

print("Intervalo de confian√ßa para Œº:", intervalo_inferior, "-", intervalo_superior)
```

**(b) Intervalo de confian√ßa de 90% para œÉ¬≤:**

Podemos calcular o intervalo de confian√ßa para a vari√¢ncia usando a seguinte f√≥rmula:

    Intervalo de confian√ßa = [(n - 1) * varia√ß√£o_amostral / valor_cr√≠tico_inferior, (n - 1) * varia√ß√£o_amostral / valor_cr√≠tico_superior]

O valor cr√≠tico inferior e superior dependem do n√≠vel de confian√ßa desejado e do tamanho da amostra menos um. Para um n√≠vel de confian√ßa de 90% e um tamanho de amostra de 10-1=9, podemos obter os valores cr√≠ticos usando a distribui√ß√£o qui-quadrado ou a fun√ß√£o de distribui√ß√£o qui-quadrado em Python.

A varia√ß√£o amostral √© dada pela soma dos quadrados das diferen√ßas entre cada valor da amostra e a m√©dia da amostra, dividida pelo tamanho da amostra menos um.

Vamos calcular o intervalo de confian√ßa para œÉ¬≤ em Python:

```python
from scipy.stats import chi2

# C√°lculo da varia√ß√£o amostral
variancia_amostral = sum((x - media_amostra) ** 2 for x in amostra) / (tamanho_amostra - 1)

# C√°lculo dos valores cr√≠ticos
valor_critico_inferior = chi2.ppf((1 + nivel_confianca) / 2, graus_de_liberdade)
valor_critico_superior = chi2.ppf((1 - nivel_confianca) / 2, graus_de_liberdade)

# C√°lculo do intervalo de confian√ßa
intervalo_inferior = (tamanho_amostra - 1) * variancia_amostral / valor_critico_inferior
intervalo_superior = (tamanho_amostra - 1) * variancia_amostral / valor_critico_superior

print("Intervalo de confian√ßa para œÉ¬≤:", intervalo_inferior, "-", intervalo_superior)
```
