# Como lidar com dados faltantes (NaN) em um Dataset

```python
import pandas as pd

dados = pd.read_csv('/content/athlete_events.csv')
```

```python
dados.head()
```

```python
dados2 = dados.dropna()   #exclui todas as linhas com algum NaN
```

```python
dados2.head()
```

```python
dados.shape
```

```python
dados2.shape
```

```python
nulo = dados.isnull()
nulo.head(100)
```

```python
faltantes = dados.isnull().sum()  #soma quantos True existem em cada coluna
print(faltantes)
```

```python
faltantes_percent = (dados.isnull().sum() / len(dados['ID']))*100
print(faltantes_percent)
```

```python
dados['Medal'].fillna('None', inplace=True)
dados['Age'].fillna(dados['Age'].mean(), inplace=True)
dados['Height'].fillna(dados['Height'].mean(), inplace=True)
dados['Weight'].fillna(dados['Weight'].mean(), inplace=True)
```
