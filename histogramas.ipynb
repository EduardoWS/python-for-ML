# Criando Histogramas 

```python
import pandas as pd

dados = pd.read_csv('/content/athlete_events.csv')
```

```python
dados.head()
```

```python
import matplotlib.pyplot as plt     #biblioteca para gr√°ficos
```

```python
dados.hist(column='Age', bins=100)  #bins = quantas barras
plt.show()
```

```python
mulheres = dados.loc[dados['Sex']=='F']
```

```python
mulheres.head()
```

```python
mulheres.hist(column='Weight', bins=100)
plt.show()
```

```python
homens = dados.loc[dados['Sex']=='M']
```

```python
homens.head()
```

```python
homens.hist(column='Weight', bins=100)
plt.show()
```
